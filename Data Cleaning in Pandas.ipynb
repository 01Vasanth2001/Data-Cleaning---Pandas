# importing

```python
import pandas as pd
```

# Reading File

```python
df = pd.read_excel(r"C:\Users\Admin\Downloads\Customer Call List.xlsx")
df
```

# Removing Duplicates

```python
df = df.drop_duplicates()
df
```

# Removing column that is useless

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

# Removing unnecessary string chars in "last name" col

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

# Similar Formatting Phone Numbers

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','')
df["Phone_Number"] = df["Phone_Number"].str.replace('nan','')
df["Phone_Number"] = df["Phone_Number"].str.replace('/','')
df
```

Converting Data Types from numbers to strings

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x:str(x))
df
```

Inserting '-' inbetween numbers in their preferred position using lambda function

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '' + x[3:6] + '' + x[6:10])
df
```

# splitting address column by no. , state, zip-code

```python
df["Address"].str.split(',',expand = True)
```

# replacing 'yes' -> y and 'no' - n

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

# Replacing 'N/A' with  ' '  in all columns

```python
df = df.replace('N/a','')
df
```

# Replacing all null values with a blank

```python
df = df.fillna('')
df
```

# CASE STUDY:

The Call Center only wants the data of people having phone numbers and those who they CAN contact/call

drop based on index

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y': # here df.loc implies that filtering is done based on index
        df.drop(x, inplace = True)
        
df        
```

we also do not want the people who didn't provide any phone numbers

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '': # here df.loc implies that filtering is done based on index
        df.drop(x, inplace = True)
        
df   
```

# Resetting Index

```python
df.reset_index(drop = True)
```

# Exporting As a CSV/Excel File

```python
df.to_excel(r'C:\Users\Admin\Downloads\Customer_Call_List.xlsx', index =False)
```
